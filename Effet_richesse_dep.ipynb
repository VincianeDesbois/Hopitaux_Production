Le premier quartile des revenus fiscaux moyens par mois est à 9471, et le troisième quartile à 11536. On sépare en deux bases : celles avec les départements avec les plus faibles revenus, et celles avec les départements avec les plus hauts revenus.

```python
# Les établissements avec des revenus départements  trés élevés

VAR_CTRL_ELEVES = VAR_CTRL.copy()
VAR_CTRL_ELEVES= VAR_CTRL_ELEVES.rename(columns = {'Moyenne Revenus Fiscaux Départementaux': 'Rev_dep'})

VAR_CTRL_ELEVES= VAR_CTRL_ELEVES[VAR_CTRL_ELEVES.Rev_dep >= 11536]
```

```python
# Les établissements avec des revenus départements trésfaibles

VAR_CTRL_FAIBLES = VAR_CTRL.copy()
VAR_CTRL_FAIBLES = VAR_CTRL_FAIBLES.rename(columns = {'Moyenne Revenus Fiscaux Départementaux': 'Rev_dep'})

VAR_CTRL_FAIBLES = VAR_CTRL_FAIBLES[VAR_CTRL_FAIBLES.Rev_dep <= 9471]
```

```python
VAR_CTRL_FAIBLES= VAR_CTRL_FAIBLES.fillna(0)
VAR_CTRL_ELEVES= VAR_CTRL_ELEVES.fillna(0)
#stjr_1 = stjr_1.fillna(0)
```
